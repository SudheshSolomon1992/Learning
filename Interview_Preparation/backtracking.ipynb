#### Combinations (LEETCODE 77) - MEDIUM

Given two integers n and k, return all possible combinations of k numbers chosen from the range [1, n].

You may return the answer in any order.

```python
def combinations(n, k):
    ans = []
    def is_valid_state(state):
        if len(state) == 2:
            return True
    
    def get_candidates(state, start):
        need = k - len(state)
        remain = n - start + 1
        available = remain - need

        for i in range(start, start + available + 1):
            yield i
        
    def backtrack(state, start):
        if is_valid_state(state):
            ans.append(state.copy())
            return
        
        for candidate in get_candidates(state, start):
            state.append(candidate)
            print("Going to child: - The current state is:", state)
            start += 1
            backtrack(state, start)
            state.remove(candidate)
            print("Going back to parent - The current state is:", state)

    backtrack([], 1)
    return ans

n = 4
k = 2
out = combinations(n, k)
print (out)
```

#### Permutations (LEETCODE 46) - MEDIUM

Given an array nums of distinct integers, return all the possible permutations. You can return the answer in any order.

```python
ans = []
def permute(nums):
    def is_valid_state(state):
        if len(state) == len(nums):
            ans.append(state[:])
            return

    def get_candidates(state):
        for i in nums:
            if i not in state:
                yield i
    
    def backtrack(state, start):
        if is_valid_state(state):
            ans.append(state.copy())
            return

        for candidate in get_candidates(state):
            state.append(candidate)
            print("Going to child: - The current state is:", state)
            backtrack(state, nums)
            state.remove(candidate)
            print("Going back to parent - The current state is:", state)
    
    backtrack([], nums)
    return ans

nums = [1,2,3]
out = permute(nums)
print (out)
```

The same number may be chosen from candidates an unlimited number of times. Two combinations are unique if the 
frequency of at least one of the chosen numbers is different.
